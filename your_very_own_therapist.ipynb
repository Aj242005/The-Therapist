```python
from datasets import load_dataset
import re

dataset = load_dataset('fadodr/mental_health_therapy')
dataset_train = dataset['train'].shuffle(seed = 42)
dataset_test = dataset['test'].shuffle(seed = 42)
print(type(dataset_train))
def transforming_conversation(sample):
  segments = [sample['instruction'],sample['input'],sample['output']]
  reformated_segments = []
  instruction = segments[0]
  input = segments[1]
  output = segments[2]
  if segments[-1] != "":
    reformated_segments.append(f'<s>[INST]<<SYS>>{instruction}<</SYS>>{input}[/INST]{output}</s>')
  else:
    reformated_segments.append(f'<s>[INST]<<SYS>>{instruction}<</SYS>>{input}[/INST]</s>')
  return {'text': ''.join(reformated_segments)}

transformed_dataset = dataset_train.map(transforming_conversation)

```

```python
transformed_dataset.push_to_hub("your_very_own_therapist")
```
